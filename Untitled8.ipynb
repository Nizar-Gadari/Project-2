```python
df['year_month'] = df['order_date'].dt.to_period('M')
df.columns
```

```python
monthly_sales = df.groupby('year_month')['sales'].sum()
```

```python
monthly_sales.plot(kind='line', figsize = (12,3), title = 'Monthly Sales Trend')
```

```python
df['year'] = df['order_date'].dt.year
```

```python
df['month'] = df['order_date'].dt.month
```

```python
monthly_by_year = df.groupby(['year', 'month'])['sales'].sum().unstack()
```

```python
monthly_by_year.T.plot(figsize=(14, 6), title='Monthly Sales By Year', marker = 'o')
```

```python
region_yearly = df.groupby([df['order_date'].dt.to_period('Y'), 'region'])['sales'].sum().unstack()
```

```python
region_yearly.plot(kind = 'bar', figsize=(14, 6 ), title = 'Yearly Sales by Region')
```

```python
segment_yearly = df.groupby([df['order_date'].dt.to_period('Y'), 'segment'])['sales'].sum().unstack()
```

```python
segment_yearly.plot(kind = 'bar', figsize=(14, 6 ), title = 'Yearly Sales by segment')
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()
monthly_sales = monthly_sales.to_timestamp()
monthly_sales = monthly_sales.reset_index()
monthly_sales.columns = ['ds', 'y']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(monthly_sales['ds'], monthly_sales['y'], marker='o')
plt.title('Monthly Total Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
