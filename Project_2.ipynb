```python
monthly_sales = df.groupby('year_month')['sales'].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(monthly_sales, trend='add', seasonal=None).fit()
forecast = model.forecast(3)
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(monthly_sales, label='Historical Sales')
plt.plot(forecast, label='Forecasted Sales', linestyle='--', marker='o')
plt.title('Sales Forecast (Next 3 Months)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np

fitted_values = model.fittedvalues
residuals = monthly_sales - fitted_values
std_error = residuals.std()

conf_int_upper = forecast + 1.96 * std_error
conf_int_lower = forecast - 1.96 * std_error
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.plot(monthly_sales.index, monthly_sales, label='Historical Sales')

plt.plot(forecast.index, forecast, label='Forecast', color='orange', linestyle='--', marker='o')

plt.fill_between(forecast.index, conf_int_lower, conf_int_upper,
                 color='orange', alpha=0.3, label='95% Confidence Interval')

plt.title('Monthly Sales Forecast with 95% Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=monthly_sales.index,
    y=monthly_sales,
    mode='lines+markers',
    name='Historical Sales'
))

fig.add_trace(go.Scatter(
    x=forecast.index,
    y=forecast,
    mode='lines+markers',
    name='Forecast',
    line=dict(dash='dash', color='orange')
))

fig.add_trace(go.Scatter(
    x=list(forecast.index) + list(forecast.index[::-1]),
    y=list(conf_int_upper) + list(conf_int_lower[::-1]),
    fill='toself',
    fillcolor='rgba(255,165,0,0.3)',
    line=dict(color='rgba(255,255,255,0)'),
    hoverinfo="skip",
    showlegend=True,
    name='95% Confidence Interval'
))

fig.update_layout(
    title='Monthly Sales Forecast with 95% Confidence Interval',
    xaxis_title='Date',
    yaxis_title='Sales',
    hovermode='x unified',
    template='plotly_white',
    width=1000,
    height=500
)

fig.show()
```

```python
def compute_confidence_bounds(forecast, std_error, confidence_level):
    from scipy.stats import norm

    z = norm.ppf(1 - (1 - confidence_level) / 2)

    margin = z * std_error

    upper = forecast + margin
    lower = forecast - margin

    return lower, upper

std_error = (monthly_sales - model.fittedvalues).std()

ci_levels = {
    "90%": compute_confidence_bounds(forecast, std_error, 0.90),
    "95%": compute_confidence_bounds(forecast, std_error, 0.95),
    "99%": compute_confidence_bounds(forecast, std_error, 0.99)
}

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=monthly_sales.index,
    y=monthly_sales,
    name='Historical Sales',
    mode='lines+markers'
))

fig.add_trace(go.Scatter(
    x=forecast.index,
    y=forecast,
    name='Forecast',
    mode='lines+markers',
    line=dict(dash='dash', color='black')
))

colors = {
    "90%": "rgba(0, 123, 255, 0.2)",
    "95%": "rgba(255, 165, 0, 0.3)",
    "99%": "rgba(220, 53, 69, 0.2)"
}

for label, (lower, upper) in ci_levels.items():
    fig.add_trace(go.Scatter(
        x=list(forecast.index) + list(forecast.index[::-1]),
        y=list(upper) + list(lower[::-1]),
        fill='toself',
        fillcolor=colors[label],
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo='skip',
        name=f'{label} CI'
    ))

fig.update_layout(
    title='Forecast with Multiple Confidence Intervals',
    xaxis_title='Date',
    yaxis_title='Sales',
    hovermode='x unified',
    template='plotly_white',
    width=1000,
    height=500
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=monthly_sales.index,
    y=monthly_sales,
    name='Historical Sales',
    mode='lines+markers'
))

fig.add_trace(go.Scatter(
    x=forecast.index,
    y=forecast,
    name='Forecast',
    mode='lines+markers',
    line=dict(dash='dash', color='black')
))

ci_traces = []
for i, (label, (lower, upper)) in enumerate(ci_levels.items()):
    trace = go.Scatter(
        x=list(forecast.index) + list(forecast.index[::-1]),
        y=list(upper) + list(lower[::-1]),
        fill='toself',
        fillcolor=colors[label],
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo='skip',
        name=f'{label} CI',
        visible=(label == "95%")
    )
    fig.add_trace(trace)
    ci_traces.append(trace)

buttons = []
for label in ci_levels:
    visibility = [True, True] + [label == k for k in ci_levels]
    buttons.append(dict(
        label=label,
        method="update",
        args=[{"visible": visibility},
              {"title": f"Forecast with {label} Confidence Interval"}]
    ))

fig.update_layout(
    updatemenus=[dict(
        type="buttons",
        direction="right",
        buttons=buttons,
        showactive=True,
        x=0.5,
        xanchor="center",
        y=1.15,
        yanchor="top"
    )],
    title="Forecast with 95% Confidence Interval",
    xaxis_title="Date",
    yaxis_title="Sales",
    hovermode="x unified",
    template="plotly_white",
    width=1000,
    height=500
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

sales_by_year = df.groupby('year')['sales'].sum().reset_index()
sales_by_year['year'] = sales_by_year['year'].astype(str)

n_frames = 30
years = sales_by_year['year']
final_values = sales_by_year['sales']

frames = []
for i in range(1, n_frames + 1):
    fraction = i / n_frames
    frame_y = (final_values * fraction).round(2)
    frame_text = [f'{val:,.0f}' for val in frame_y]

    frames.append(go.Frame(
        data=[go.Bar(
            x=years,
            y=frame_y,
            text=frame_text,
            textposition='outside',
            marker_color='#006400'
        )],
        name=str(i)
    ))

initial_y = [0] * len(years)
initial_text = ['0'] * len(years)

fig = go.Figure(
    data=[go.Bar(
        x=years,
        y=initial_y,
        text=initial_text,
        textposition='outside',
        marker_color='#006400'
    )],
    layout=go.Layout(
        title='Total Sales by Year (Animated)',
        xaxis_title='Year',
        yaxis_title='Sales',
        yaxis_range=[0, final_values.max() * 1.1],
        bargap=0.4,
        updatemenus=[dict(type='buttons', showactive=False)],
        transition=dict(duration=0),
        sliders=[dict(visible=False)]
    ),
    frames=frames
)

fig.update_layout(
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label=None,
            method='animate',
            args=[None, {
                "frame": {"duration": 50, "redraw": True},
                "fromcurrent": True,
                "transition": {"duration": 0},
                "mode": "immediate"
            }]
        )]
    )]
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("train.csv", sep=";")

df.rename(columns={
    'order_date': 'Order Date',
    'product_name': 'Product Name',
    'year_month': 'Year Month',
    'sales': 'Sales'
}, inplace=True)

df['Order Date'] = pd.to_datetime(df['Order Date'])

if 'Year Month' not in df.columns or df['Year Month'].isnull().all():
    df['Year Month'] = df['Order Date'].dt.to_period('M').astype(str)

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()

fig1 = px.bar(top_products, x='Sales', y='Product Name', orientation='h',
              title='Top 10 Products by Sales', color_discrete_sequence=['green'])
fig1.update_layout(yaxis=dict(autorange='reversed'))
fig1.show()

monthly_sales = df.groupby('Year Month')['Sales'].sum().reset_index()

fig2 = px.line(monthly_sales, x='Year Month', y='Sales', markers=True, title='Monthly Sales Trend')
fig2.show()
```
